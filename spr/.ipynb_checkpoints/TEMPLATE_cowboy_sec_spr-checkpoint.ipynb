## Expression/purification parameters
- `oligomers`: whether or not your designs are oligomers (if `True`, a column `n_chains` needs to be specified in `echo_df`).
- `culture_vol`: culture volume, formated as a string `AxB` where `B` is the volume of one culture (in mL), and `A` the number of consolidated cultures.
- `culture_time`: time between inocculation and harvest, in hours.
- `culture_temp`: temperature during growth/expression, in degree Celsius.
- `sec_column`: column used for SEC, e.g. `S75_5-150`, `S200_5-150`,`S75_10_300`,etc.
- `sec_buffer`: buffer used for SEC.
- `sec_buffer_pH`: the pH of the SEC buffer. 
- `sec_instrument`: the instrument used for SEC. Must be either `akta` or `hplc`.
- `sec_calibration`: instrument and column-specific calibration file (`.json` format)
        - Calibrations are kept in /net/software/lab/cowboy/calibrations/
- `sec_traces`: dictionary of lists of paths to exported SEC traces. 
    - `AKTA` data structure should be `/path_to_your_run/*.zip`. The list should contain paths to the individual `.zip` files.
    - `HPLC` data structure should be `/path_to_your_run/trace_folders/*.CSV`. The list should contain paths to the individual *trace folders*, NOT the individual `.CSV` files. The folder containing the individual trace folders should also contain the `fractions.csv` file containing the mapping between traces and fractions.

```python


def write_spr_csv(spr_df, filename,dest_plate):
    user = getpass.getuser()
    filename = f"{filename}_{user}_spr.csv"
    conc_header = ""
    pos_header = ""
    dilute_string = ""
    dilution = spr_df.normed_conc_nM[0]
    for i in range(spr_df.dilutions[0]):
        conc_header += f",Concentration {i+1}"
        pos_header += f",Position {i+1}"
        dilute_string = f',{round(dilution)}' + dilute_string
        dilution /= spr_df.dilution_step[0]
    header = "plate id,solution" + conc_header + pos_header + "\n"
    if dest_plate == 384:
        plate_format = [[r + str(c) for c in range(1,25)] for r in 'ACEGIKMNOBDFHJLNP']
    if dest_plate == 96:
        plate_format = [[r + str(c) for c in range(1,13)] for r in 'ABCDEFG']
    with open(filename, "w") as file:
        file.write(header)
    f = open(filename, "a")
    well = 0
    row = 0
    plate = 1
    plate_array = []
    well_array = []

    for w, r in spr_df.set_index("well_id").iterrows():
        pos_string = ""
        well = int(np.floor(((row+1)/len(plate_format))))*r.dilutions
        #print (f'{plate}_{plate_format[row%len(plate_format)][well]}')
        plate_array.append(plate)
        well_array.append(plate_format[row%len(plate_format)][well])
        for i in range(r.dilutions):
            pos_string = f",{plate_format[row%len(plate_format)][well]}" + pos_string
            well += 1
        row += 1
        if well >= int(dest_plate):
            plate += 1
            well = 0
        line= f"{plate},{w}"+ dilute_string + pos_string + "\n"
        f.write(line)
    spr_df['Destination Plate Name'] = plate_array
    spr_df['Destination Well'] = well_array
    return spr_df
```

```python
#generate csv for spr
spr_df = write_spr_csv(spr_df,"test_def_spr", plate)
```

```python
def gen_spr_ot2_script(
    df,
    pool_plate = "384",
    filename="",
    path="",
    date=today,
    template="/net/software/lab/cowboy/SPR_norm_and_dilute.py",
):
    """
    Generate OT-2 script for pooling and normalising SEC fractions.

    df: dataframe containing the experimental data.
    
    pool_plate: type of plate to pool into, "384" or "96"
    
    path: a path to prepend onto the script output will be in format f"{path}{date}_{filename}{user}_spr_norm_and_dilute.py"

    filename: filename for the generated script. output will be in format f"{path}{date}_{filename}{user}_spr_norm_and_dilute.py"

    date: string of the date that gets prepended to the output files.

    template: path to an OT-2 protocol template.
    """
    
    user = getpass.getuser()
    file_name = f"{path}{date}_{filename}{user}_spr_norm_and_dilute"

    ot2_transfers = ""
    non_normed_transfers = ""
    for i, r in df.iterrows():
        volume = round(r.pooled_vol_corr,6)
        pooled_frac_list = " ".join(
            [
                f"'{a}',"
                for a in str(r.well_id)
                .replace("['", "")
                .replace("']", "")
                .split("' '")
                if a != ""
            ]
        )

        pooled_frac_vol_list = [
            float(a) * 1e3
            for a in str(volume).replace("[", "").replace("]", "").split(" ")
            if a != ""
        ]
        buffer_vol = r.buffer_vol * 1e3
        writestr = f"[ [{pooled_frac_list}], {pooled_frac_vol_list}, {r['Destination Plate Name']}, '{r['Destination Well']}', {buffer_vol:3.3f}],\n"
        ot2_transfers += writestr
    ot2_template = open(template, "r").readlines()

    with open(file_name + ".py", "w") as f:
        for l in ot2_template:
            modified_line = l.replace("### PROTOCOL_NAME ###", file_name)
            modified_line = modified_line.replace("### USER_NAME ###", user)
            modified_line = modified_line.replace("### POOL_PLATE ###", str(pool_plate))
            modified_line = modified_line.replace("### TRANSFERS ###", ot2_transfers)
            modified_line = modified_line.replace('### DILUTIONS ###', str(df.dilutions[0]))
            modified_line = modified_line.replace("### STEP ###", str(df.dilution_step[0]))
            modified_line = modified_line.replace("### VOLUME ###", str(df.volume_step[0]))
            f.write(modified_line)
    return ot2_transfers
```
